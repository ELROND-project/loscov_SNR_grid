The goal of this notebook is to explicitely compute the LOS shear - galaxy shape covariance. Namely, we want to compute 

$$ \mathrm{Cov}\Big[\hat{\xi}_{\gamma_{\rm LOS}\times\varepsilon}^\eta(a,b), \hat{\xi}_{\gamma_{\rm LOS}\times\varepsilon}^\eta(a', b') \Big] = \Big\langle\hat{\xi}_{\gamma_{\rm LOS}\times\varepsilon}^\eta(a,b) \hat{\xi}_{\gamma_{\rm LOS}\times\varepsilon}^\eta(a', b') \Big\rangle  - \Big\langle \hat{\xi}_{\gamma_{\rm LOS}\times\varepsilon}^\eta(a,b) \Big\rangle \Big\langle \hat{\xi}_{\gamma_{\rm LOS}\times\varepsilon}^\eta(a', b') \Big\rangle$$

where $\eta = \pm 1$.

The notebook is structured as follows:

In [Preliminaries](#preliminaries), we set up all the background stuff we need to carry out our calculations. First, we [import](#packages) all the necessary packages, and define some [useful functions]($usefulfunctions). We then [define our cosmology](#cosmology_and_weyl_power_spectrum) in CAMB, and use this to generate the Weyl power spectrum from which we can calculate quantities of interest. [Finally](#distributions), we import the forecasted sample of Euclid lenses, extract their redshift distributions, and create PDFs from these distributions. We also define the distribution of galaxies we expect to observe.

Next, we calculate the [autocorrelation function](#LOS_shear_autocorrelation_function) for $\gamma_\mathrm{LOS}$. There are several steps to this. First, we must [define the LOS weight function](#weight_functions_shear). Using these weight functions, we can calculate the (redshift-averaged) [angular power spectra](#angular_power_spectrum_shear) for $\gamma_\mathrm{LOS}$. With these angular power spectra, we can calculate the actual [autcorrelation functions](#correlation_function_shear) for $\gamma_\mathrm{LOS}$ themselves, as integrals of these power spectra.

We then repeat this process for the [autocorrelation function](#shape_autocorrelation_function) for $\gamma$, the weak lensing shear. We once again start by defining the relevant [weight function](#weight_function_shape), from which we calculate the (redshift-averaged) [angular power spectra](#angular_power_spectrum_shape) for $\gamma$. This allows us to calculate the [autocorrelation function](#correlation_function_shape) for the weak lensing shear.

The last correlations to calculate are the [cross correlation functions](#gamma_shape_correlation_function) for $\gamma_\mathrm{LOS}\times\gamma$. We calculate the [cross power spectrum](#gamma_shape_power_spectrum), and use this to determine the [correlation functions](#correlation_function_shear_shape).

With the correlation functions ready, we can proceed to [calculating the covariance matrix](#covariance_calculation) we need. We split this into three sections, starting with the [Noise covariance](#noise_variance), then the [Cosmic covariance](#cosmic_variance), and finally the [sparsity covariance](#sparsity_variance).

### 5.1 Noise covariance <a id='noise_covariance'></a>

```python
import pickle

def read_matrices():
    """
    Reads NCOV, CCOV, and SCOV matrices for all combinations of b1 and b2 (from 1 to 5).
    Returns three nested dictionaries containing all the matrices.
    """
    NCOV = {}
    CCOV = {}
    SCOV = {}
    
    for b1 in range(1, 6):
        
        try:
            with open(f"Matrices/LLLe/ncov{b1}", "rb") as f:
                NCOV[b1-1] = pickle.load(f)
        except Exception as ex:
            print(f"Error reading NCOV for b1={b1}: {ex}")

        try:
            with open(f"Matrices/LLLe/ccov{b1}", "rb") as f:
                CCOV[b1-1] = pickle.load(f)
        except Exception as ex:
            print(f"Error reading CCOV for b1={b1}: {ex}")

        try:
            with open(f"Matrices/LLLe/scov{b1}", "rb") as f:
                SCOV[b1-1] = pickle.load(f)
        except Exception as ex:
            print(f"Error reading SCOV for b1={b1}: {ex}")

    return NCOV, CCOV, SCOV, xi_plus, xi_minus

NCOV, CCOV, SCOV, xi_plus, xi_minus = read_matrices()
```

```python
title = 'Noise covariance (1e-14)'
cmap="viridis"
use_standardised_range = False

# Determine global vmin and vmax
all_values = [
    np.array(matrix['full']).flatten() * 1e14
    for b1, matrix in NCOV.items()
    if 'full' in matrix
]
if all_values:
    global_min = min(arr.min() for arr in all_values)
    global_max = max(arr.max() for arr in all_values)
else:
    raise ValueError("No valid matrices found in NCOV.")

fig, axes = plt.subplots(1, 5, figsize=(15, 3.5))
fig.suptitle(title, fontsize=16)

for b1 in range(1, 6):
    ax = axes[b1-1]  # Access the subplot at position (b1-1, b2-1)
    matrix = NCOV[b1-1].get("full")  # Get the matrix using the provided key
            
    if matrix is not None:
        if use_standardised_range:
            im = ax.imshow(matrix * 1e14, cmap=cmap, aspect="auto", vmin=global_min, vmax=global_max)
        else:
            im = ax.imshow(matrix * 1e14, cmap=cmap, aspect="auto")
                
        plt.colorbar(im, ax=ax, shrink=0.7)
        ax.set_title(f"b1={b1}", fontsize=10)
    else:
        ax.axis("off")  # Turn off the axis if no data is available

    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout(rect=[0, 0, 1, 0.99])  # Leave space for the suptitle
plt.savefig('Noise_covariance.pdf')
plt.show()
```
